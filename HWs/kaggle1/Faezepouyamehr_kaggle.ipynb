```python
def generte_train_test( n_train, n_valid, train_eile='drive/MyDrive/train.csv', test_file='drive/MyDrive/test.csv', label_file='drive/MyDrive/train_result.csv'):
  # data cleaning
  null_col = df.isnull().sum()
  remove = []
  for i in range(len(null_col)):
      if null_col[i]!=0:
          remove.append(null_col.index[i])
  df.drop(remove, inplace =True, axis =1)
  null_col = df_test.isnull().sum()
  remove = []
  for i in range(len(null_col)):
      if null_col[i]!=0:
          remove.append(null_col.index[i])

  df_test.drop(remove, inplace =True, axis =1)
  df_n=df.to_numpy()
  df_n_test = df_test.to_numpy()

  y_class= y["Class"].to_numpy()

  inds = np.arange(df_n.shape[0])
  np.random.shuffle(inds)
  train_inds = inds[:n_train]
  valid_inds = inds[n_train:n_train+n_valid]
  X_trainset = df_n[train_inds] 
  Y_trainset=y_class[train_inds]
  X_valid = df_n[valid_inds] 
  Y_valid=y_class[valid_inds]

  X_testset= df_n_test

  return X_trainset,Y_trainset, X_valid, Y_valid, X_testset
```

```python
class MYLOGREG:
        
    def fit(self,C, X, y, X_valid, y_valid, init_weights, alpha=0.2, lam= 0.01, epochs=1000):
        self.alpha=alpha
        self.lam=lam
        self.N = X.shape[0]
        self.num_classes = np.arange(19)
        self.loss_train = [0 for i in range(epochs)]
        self.loss_valid = [0 for i in range(epochs)]
        self.acc_train = [0 for i in range(epochs)]
        self.acc_valid = [0 for i in range(epochs)]
        
        X = np.insert(X, 0, 1, axis=1)
        X_valid = np.insert(X_valid, 0, 1, axis=1)
        y= self.one_hot(y)
        y_valid = self.one_hot(y_valid)
        
        # self.weights = np.random.zeros(len(self.num_classes), X.shape[1])
        if init_weights is not None:
          self.weights = init_weights
        else:
          self.weights = np.zeros((len(self.num_classes), X.shape[1]))
        
        for i in range(epochs):          
          predicted = self.predict_prob(X)
          predicted_v = self.predict_prob(X_valid)
          self.loss_train[i]=self.cross_entropy_loss(y, predicted)
          self.loss_valid[i]=self.cross_entropy_loss(y_valid, predicted_v)
          
          P = self.predict_prob(X)
          gradient = X.T.dot(P - y)/X.shape[0] +  (self.lam/self.N)*self.weights.T
          self.weights -= self.alpha * gradient.T
          print('At iterations: ', i+1, ' training acccuracy is: ', self.score_(X, y))
          print('At iterations: ', i+1, ' valid acccuracy is: ', self.score_(X_valid, y_valid))
          self.acc_train[i] = self.score_(X, y)
          self.acc_valid[i] = self.score_(X_valid, y_valid)
        return self

    def cross_entropy_loss(self, y, probs):
      out1= (-1 / len(probs)) * (y* np.log(probs)).sum()
      out2= (self.lam/(2.0*self.N))*np.linalg.norm(self.weights)**2
      return out1+ out2

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        return self.predict_prob(X)
    
    def predict_prob(self, X):
        prob_vals = np.dot(X, self.weights.T).reshape(-1,len(self.num_classes))
        return self.softmax(prob_vals)
    
    def softmax(self, z):
        return np.exp(z) / np.sum(np.exp(z), axis=1).reshape(-1,1)

    def predict_classes(self, X):
        return np.argmax(self.predict(X), axis=1)
    
    def one_hot(self, y):
        class_labels = {c:i for i,c in enumerate(self.num_classes)}
        return np.eye(len(self.num_classes))[np.vectorize(lambda c: class_labels[c])(y).reshape(-1)]

    def score(self, X, y):
        return np.mean(self.predict_classes(X) == y)    

    def score_(self, X, y):
        probs = self.predict_prob(X)
        return np.mean(np.argmax(probs, axis=1) == np.argmax(y, axis=1))
```
